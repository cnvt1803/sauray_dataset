```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')

model.train(
    data='/content/drive/MyDrive/dataYolo/data.yaml',
    epochs=50,
    imgsz=640,
    batch=4
)
```

```python
metrics = model.val()
print(metrics)
```
